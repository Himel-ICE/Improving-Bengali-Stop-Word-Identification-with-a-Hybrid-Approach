```python
def clean_text(text):
    text = emoji.replace_emoji(text, replace='')
    text = re.sub(r'[a-zA-Z]', '', text)
    text = re.sub(r'[,.?]', '', text)

    return text
```

```python
df_filtered.head['tokenized_comment'] = df_filtered.head['comment'].apply(lambda x: x.split())
```

```python
stopword_list = list(stopwords_pos.keys())

def filter_tokens(tokens):
    without_stopword = [word for word in tokens if word not in stopword_list]
    only_stopwords = [word for word in tokens if word in stopword_list]
    return pd.Series([without_stopword, only_stopwords])
```

```python
df_filtered[['comment_without_stopword', 'stopword']] = df_filtered['tokenized_comment'].apply(filter_tokens)
```

```python
dataset = df_filtered[['sentence', 'tokenized_comment', 'comment_without_stopword', 'stopword']]
```
